```python
from autogluon.tabular import TabularDataset, TabularPredictor
import numpy as np
```

```python
train_data = TabularDataset('./data/raw/training.csv')
train_data
```

```python
import os
# force the text predictor (enabled by hyperparameters='multimodal') to train text model on CPU
os.environ["AUTOGLUON_TEXT_TRAIN_WITHOUT_GPU"] = "1"
# time limit (seconds) specifies approx. how long fit should run
predictor = TabularPredictor(label='Unwanted', path='models', verbosity=3)\
.fit(train_data, time_limit=None, presets='best_quality',
     hyperparameters='multimodal'
    )
```

```python
# see what models are fitted
results = predictor.fit_summary()
```

```python
validation_data = TabularDataset('./data/raw/valid.csv')
test_data = TabularDataset('./data/raw/test.csv')
```

```python
valid_pred = predictor.predict(validation_data)
valid_pred
# np.savetxt("valid.txt", valid_pred, fmt='%s')
```

```python
test_pred = predictor.predict(test_data)
test_pred
```

```python
import os
sub_name = "submission_autogluon_baseline"
os.makedirs(sub_name, exist_ok=True)
np.savetxt(sub_name+"/"+"valid.txt", valid_pred, fmt='%s')
np.savetxt(sub_name+"/"+"test.txt", test_pred, fmt='%s')


import os
os.system("cd {} &&zip {}.zip test.txt valid.txt".format(sub_name, sub_name))
```
