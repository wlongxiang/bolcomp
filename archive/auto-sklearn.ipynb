```python
import autosklearn.classification
import numpy as np

import pandas as pd
```

```python
train_df = pd.read_csv("./data/raw/training.csv")
eval_df = pd.read_csv("./data/raw/valid.csv")
test_df = pd.read_csv("./data/raw/test.csv")
```

```python
# drop label and unneeded columns to form trainX
X_train = train_df[train_df.columns.drop(["Unwanted", "GlobalID", "image_url"])]
y_train = train_df["Unwanted"]
X_train.sample(20)
```

```python
# it does seem that it is more likely for rows without title to be unwanted
```

```python
# there are rows without product title
df_no_title = train_df[X_train["ProductTitleNL"].isnull()]
df_no_title
count_wanted = df_no_title[df_no_title.Unwanted == "Y"].shape[0]
count_unwanted = df_no_title[df_no_title.Unwanted == "N"].shape[0]
print("unwanted percent is:",(count_unwanted)/(count_unwanted+ count_wanted) )
```

```python
# there are rows without description title
df_no_desc = train_df[X_train["ProductDescriptionNL"].isnull()]
df_no_desc
count_wanted = df_no_desc[df_no_desc.Unwanted == "Y"].shape[0]
count_unwanted = df_no_desc[df_no_desc.Unwanted == "N"].shape[0]
print("unwanted percent is:",(count_unwanted)/(count_unwanted+ count_wanted) )
```
